```python
import matplotlib.pyplot as plt
import numpy as np

# Data extracted from the benchmarks
videos = ['v.mp4', 'v3.mp4', 'v2.mp4', 'v4.mp4', 'v6.mp4', 'v7.mp4', 'v8.mp4']
yolov5n_fps = [35.12, 35.76, 32.41, 35.55, 33.56, 38.06, 33.19]
yolov5s_fps = [19.86, 21.19, 19.55, 19.89, 20.21, 20.32, 20.27]
yolov5m_fps = [10.41, 9.20, 9.99, 10.50, 10.20, 11.17, 10.11]

# Plotting
x = np.arange(len(videos))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, yolov5n_fps, width, label='YOLOv5n')
rects2 = ax.bar(x, yolov5s_fps, width, label='YOLOv5s')
rects3 = ax.bar(x + width, yolov5m_fps, width, label='YOLOv5m')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Videos')
ax.set_ylabel('Average FPS')
ax.set_title('YOLOv5 Model Performance on Different Videos')
ax.set_xticks(x)
ax.set_xticklabels(videos)
ax.legend()

# Function to add labels on top of bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data extracted from the benchmarks
videos = ['v.mp4', 'v3.mp4', 'v2.mp4', 'v4.mp4', 'v6.mp4', 'v7.mp4', 'v8.mp4']
yolov5n_fps = [35.12, 35.76, 32.41, 35.55, 33.56, 38.06, 33.19]
yolov5s_fps = [19.86, 21.19, 19.55, 19.89, 20.21, 20.32, 20.27]
yolov5m_fps = [10.41, 9.20, 9.99, 10.50, 10.20, 11.17, 10.11]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(videos, yolov5n_fps, marker='o', label='YOLOv5n')
plt.plot(videos, yolov5s_fps, marker='o', label='YOLOv5s')
plt.plot(videos, yolov5m_fps, marker='o', label='YOLOv5m')

# Annotate FPS values
for i, fps in enumerate(yolov5n_fps):
    plt.annotate(f'{fps:.2f}', (videos[i], yolov5n_fps[i]), textcoords="offset points", xytext=(0,10), ha='center')
for i, fps in enumerate(yolov5s_fps):
    plt.annotate(f'{fps:.2f}', (videos[i], yolov5s_fps[i]), textcoords="offset points", xytext=(0,10), ha='center')
for i, fps in enumerate(yolov5m_fps):
    plt.annotate(f'{fps:.2f}', (videos[i], yolov5m_fps[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.xlabel('Videos')
plt.ylabel('Average FPS')
plt.title('Benchmarking through 7 testing videos')
plt.xticks(rotation=45)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data preparation
data = {
    'Video': ['v.mp4', 'v3.mp4', 'v2.mp4', 'v4.mp4', 'v6.mp4', 'v7.mp4', 'v8.mp4'],
    'YOLOv5n': [35.12, 35.76, 32.41, 35.55, 33.56, 38.06, 33.19],
    'YOLOv5s': [19.86, 21.19, 19.55, 19.89, 20.21, 20.32, 20.27],
    'YOLOv5m': [10.41, 9.20, 9.99, 10.50, 10.20, 11.17, 10.11]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)
df.set_index('Video', inplace=True)

# Create a heat map
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('YOLOv5 Model Performance (Average FPS Heat Map)')
plt.xlabel('Model')
plt.ylabel('Video')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
